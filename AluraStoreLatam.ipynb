### Importación de datos

#1. Análisis de facturación



```python
ingresos_tienda1 = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso Total Tienda 1: ${ingresos_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingresos_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingresos_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingresos_tienda4:,.2f}")
```

```python
tiendas = {
    'tienda': ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4'],
    'ingresos': [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]
}
df_ingresostotal = pd.DataFrame(tiendas).set_index('tienda')
df_ingresostotal.plot(
    kind='barh',
    figsize=(12, 8),
    color='purple',
    legend=False,
    title='Ingresos Totales por Tiendas'
)
```
