#1. Análisis de facturación



```python
ingresos_tienda1 = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso Total Tienda 1: ${ingresos_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingresos_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingresos_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingresos_tienda4:,.2f}")
```

```python
tiendas = {
    'tienda': ['tienda 1', 'tienda 2', 'tienda 3', 'tienda 4'],
    'ingresos': [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]
}
df_ingresostotal = pd.DataFrame(tiendas).set_index('tienda')
df_ingresostotal.plot(
    kind='barh',
    figsize=(12, 8),
    color='purple',
    legend=False,
    title='Ingresos Totales por Tiendas'
)
```

# 2. Ventas por categoría

```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

todas_las_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])
df_popularidad = todas_las_tiendas.groupby(['Categoría del Producto', 'Tienda']).size().unstack().fillna(0)
df_popularidad['Total Global'] = df_popularidad.sum(axis=1)
df_top_popular = df_popularidad.sort_values('Total Global', ascending=True).tail(5)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.heatmap(df_top_popular.drop(columns='Total Global'),
            annot=True,
            fmt=".0f",
            cmap='Blues',
            linewidths=.5)

plt.title('Intensidad de Ventas por Categoría y Tienda')
plt.show()
```

# 4. Productos más y menos vendidos

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos_extremos = []

for i in range(len(tiendas)):

    conteo_productos = tiendas[i]['Producto'].value_counts()

    mas_vendido = conteo_productos.idxmax()
    cant_max = conteo_productos.max()
    menos_vendido = conteo_productos.idxmin()
    cant_min = conteo_productos.min()

    datos_extremos.append({
        'Tienda': tiendas[i],
        'Producto Estrella': mas_vendido,
        'Ventas Max': cant_max,
        'Producto Rezagado': menos_vendido,
        'Ventas Min': cant_min
    })
df_resumen_extremos = pd.DataFrame(datos_extremos)
display(df_resumen_extremos)
```
